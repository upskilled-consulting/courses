```python
# You can use this to convert a .ply file to a .splat file programmatically in python
# Alternatively you can drag and drop a .ply file into the viewer at https://antimatter15.com/splat

from plyfile import PlyData
import numpy as np
import argparse
from io import BytesIO


def process_ply_to_splat(ply_file_path):
    plydata = PlyData.read(ply_file_path)
    vert = plydata["vertex"]
    sorted_indices = np.argsort(
        -np.exp(vert["scale_0"] + vert["scale_1"] + vert["scale_2"])
        / (1 + np.exp(-vert["opacity"]))
    )
    buffer = BytesIO()
    for idx in sorted_indices:
        v = plydata["vertex"][idx]
        position = np.array([v["x"], v["y"], v["z"]], dtype=np.float32)
        scales = np.exp(
            np.array(
                [v["scale_0"], v["scale_1"], v["scale_2"]],
                dtype=np.float32,
            )
        )
        rot = np.array(
            [v["rot_0"], v["rot_1"], v["rot_2"], v["rot_3"]],
            dtype=np.float32,
        )
        SH_C0 = 0.28209479177387814
        color = np.array(
            [
                0.5 + SH_C0 * v["f_dc_0"],
                0.5 + SH_C0 * v["f_dc_1"],
                0.5 + SH_C0 * v["f_dc_2"],
                1 / (1 + np.exp(-v["opacity"])),
            ]
        )
        buffer.write(position.tobytes())
        buffer.write(scales.tobytes())
        buffer.write((color * 255).clip(0, 255).astype(np.uint8).tobytes())
        buffer.write(
            ((rot / np.linalg.norm(rot)) * 128 + 128)
            .clip(0, 255)
            .astype(np.uint8)
            .tobytes()
        )

    return buffer.getvalue()


def save_splat_file(splat_data, output_path):
    with open(output_path, "wb") as f:
        f.write(splat_data)
```

```python
import os
from datetime import datetime

# Path to your working directory
working_dir = './'

# List all subdirectories that match the YYYY-MM-DD format
date_folders = []
for name in os.listdir(working_dir):
    folder_path = os.path.join(working_dir, name)
    if os.path.isdir(folder_path):
        try:
            date_obj = datetime.strptime(name, "%Y-%m-%d")
            date_folders.append((date_obj, name))
        except ValueError:
            continue  # Skip non-date folder names

# Get the folder with the most recent date
if date_folders:
    latest_folder = max(date_folders, key=lambda x: x[0])[1]
    print("Most recent folder:", latest_folder)
else:
    print("No valid date-named folders found.")
```

```python
import os
import re
from datetime import time

# Path to your working directory
working_dir = latest_folder

# Regex to find time strings in folder names (HH-MM-SS)
time_pattern = re.compile(r'\b(\d{2})-(\d{2})-(\d{2})\b')

most_recent = None
most_recent_folder = None

for name in os.listdir(working_dir):
    folder_path = os.path.join(working_dir, name)
    if os.path.isdir(folder_path):
        match = time_pattern.search(name)
        if match:
            try:
                t = time(int(match.group(1)), int(match.group(2)), int(match.group(3)))
                if not most_recent or t > most_recent:
                    most_recent = t
                    most_recent_folder = name
            except ValueError:
                continue  # Skip invalid time values

if most_recent_folder:
    print("Most recent time-containing folder:", most_recent_folder)
else:
    print("No valid time-containing folders found.")
```

```python
# Path to the 'output' folder
parent_folder = latest_folder
project = "/bamburgh-castle"

input_file = f'./{parent_folder}/{most_recent_folder}/point_cloud/iteration_30000/point_cloud.ply'
output_folder = '/'.join(input_file.split('/')[0:2])
output_file =  output_folder + project + '.splat'

splat_data = process_ply_to_splat(input_file)

save_splat_file(splat_data, output_file)
```
